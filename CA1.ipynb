```python
#barplot
df['eventDate'] = pd.to_datetime(df['eventDate'])  # Ensure eventDate is in datetime format
df['year'].value_counts().sort_index().plot(kind='bar')
plt.show()
```

```python
# Group the data by 'species' and 'countryCode', then count the occurrences
species_country_count = df.groupby(['countryCode', 'species']).size().reset_index(name='count')
species_country_count
# Display the first few rows of the result
#print(species_country_count.head())
```

```python
# Assuming df is your DataFrame
species_per_country = df.groupby('countryCode')['species'].nunique().reset_index(name='species_count')
max_species_country = species_per_country.loc[species_per_country['species_count'].idxmax()]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='countryCode', y='species_count', data=species_per_country)
plt.axhline(max_species_country['species_count'], color='red', linestyle='--', label='Max Species Count')
plt.title('Unique Species Count by Country')
plt.xlabel('Country Code')
plt.ylabel('Unique Species Count')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Group the data by 'countryCode' and count distinct species in each country
species_per_country = df.groupby('countryCode')['species'].nunique().reset_index(name='species_count')

# Find the country with the maximum number of species
max_species_country = species_per_country.loc[species_per_country['species_count'].idxmax()]
species_per_country
max_species_country
```

```python
 

# Assuming df is your DataFrame
genus_per_country = df.groupby('countryCode')['genus'].nunique().reset_index(name='genus_count')
max_genus_country = genus_per_country.loc[genus_per_country['genus_count'].idxmax()]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='countryCode', y='genus_count', data=genus_per_country)
plt.axhline(max_genus_country['genus_count'], color='red', linestyle='--', label='Max Genus Count')
plt.title('Unique Genus Count by Country')
plt.xlabel('Country Code')
plt.ylabel('Unique Genus Count')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
